```python
import matplotlib.pyplot as pit
```

```python
pit.figure()
pit.plot([1,2,3,4,5,6,7,8,9,8,7,6,5,4,3,2,1,0])
pit.show()
```

```python
import numpy as np
t = np.arange(0,12,0.01)
y = np.sin(t)
```

```python
pit.figure(figsize=(10,6))
pit.plot(t,y)
pit.show()
```

```python
pit.figure(figsize=(10,6))
pit.plot(t,y)
pit.grid()
pit.xlabel('time')
pit.ylabel('Amplitude')
pit.title('Example of sinewave')
pit.show()
```

```python
pit.figure(figsize=(10,6))
pit.plot(t, np.sin(t))
pit.plot(t, np.cos(t))
pit.grid()
pit.xlabel('time')
pit.ylabel('Amplitude')
pit.title('Example of sinewave')
pit.show()
```

```python
pit.figure(figsize=(10,6))
pit.plot(t, np.sin(t), label='sin')
pit.plot(t, np.cos(t), label='cos')
pit.grid()
pit.legend()
pit.xlabel('time')
pit.ylabel('Amplitude')
pit.title('Example of sinewave')
pit.show()
```

```python
pit.figure(figsize=(10,6))
pit.plot(t, np.sin(t), lw=3, label='sin')
pit.plot(t, np.cos(t), 'r', label='cos')
pit.grid()
pit.legend()
pit.xlabel('time')
pit.ylabel('Amplitude')
pit.title('Example of sinewave')
pit.show()
```

```python
s1 = np.random.normal(loc=0, scale=1, size=1000)
s2 = np.random.normal(loc=5, scale=0.5, size=1000)
s3 = np.random.normal(loc=10, scale=2, size=1000)
```

```python
pit.figure(figsize=(10,6))
pit.plot(s1, label='s1')
pit.plot(s2, label='s2')
pit.plot(s3, label='s3')
pit.legend()
pit.show()
```

```python
pit.figure(figsize=(10,6))
pit.boxplot((s1, s2, s3))
pit.grid()
pit.show()
```
